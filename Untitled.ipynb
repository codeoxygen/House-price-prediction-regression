# Data Cleaning

```python
#function to get the number of  missing values and it's percentage
#feature names are arranged in descending order(from the maximum number of amount to the smallest) 
def missing_data(df):
    total = df.isnull().sum().sort_values(ascending = False)
    Percentage = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
    
    return pd.concat([total, Percentage], axis=1, keys=['Total','Percentage'])
missing = missing_data(train)
missing.head(30)#display only  the firs 30 features with the highest percentage
```

```python
#drop the entire features that have more than 15% missing values. Because filling over 40% values in manually is not accurate
drop_features = list(missing[missing.Percentage>15].index) # Feature4s to be dropped
drop_features
```

```python
train.drop(columns=drop_features,axis=0,inplace=True) # drop those columns
```

```python
#features to be cleaned
print(missing[(missing['Percentage']>0) & (missing['Percentage']<15)])
cleaning_features = list(missing[(missing['Percentage']>0) & (missing['Percentage']<15)].index)
print('Features needs to be cleaned :', '\n', cleaning_features)
```

```python
cleaning_df = train[cleaning_features]
cleaning_df.head()
```

```python
cat_features = [i for i in train.columns if train[i].dtypes=='O'] #categoricol features
```

```python
cleaning_df.info()
```

```python
cat_cols = ['GarageCond', 'GarageType', 'GarageFinish', 'GarageQual', 'BsmtFinType2', 'BsmtExposure', 'BsmtQual', 'BsmtCond', 'BsmtFinType1', 'MasVnrType', 'Electrical']
# identifying the unique categorical values in each these features. We can confirm which type of missing values are included here
for i in cat_cols:
    print(str(i), ': ',train[i].unique())
```

```python
#functions to clean the dataset
def filling_categorical(features,df):#filling categorical values with it's mode
    for i in features:
        df[i].fillna(df[i].mode()[0],inplace =True)
    return df

def filling_year(features,df):# filling the year with it's previous value
    for i in features:
        df[i].fillna(method='ffill',inplace = True)
    return df
def filling_numerical(features,df):
    for i in features:
        df[i].fillna(df[i].mean(),inplace =True)
    return df
    
```

```python
train = filling_categorical(cat_cols,train) # filling categorical values
train = filling_year(['GarageYrBlt'],train) #filling year values
train = filling_numerical(['MasVnrArea'],train) # filling numerical values
```

# Pre-processing
### Categorical feature encoding

```python
train.Exterior2nd.value_counts().to_dict()
```

```python
# Neighborhood,Exterior1st,Exterior2nd have more categorical values (cell [178])
#I'm gonna use frequency count encoding method for these 3 columns
neighb_freq =train.Neighborhood.value_counts().to_dict() # get the frequency of each categories in stored the value ahead of the caregory name.
train.Neighborhood = train.Neighborhood.map(neighb_freq)# replace the relevant category with it's freq value

#repeat the process to other columns
ex1st_freq=train.Exterior1st.value_counts().to_dict()
train.Exterior1st = train.Exterior1st.map(ex1st_freq)

ex2nd_freq =train.Exterior2nd.value_counts().to_dict()
train.Exterior2nd =train.Exterior2nd.map(ex2nd_freq)
```

```python
#let's encode ordinal variables with label encoding and nominal variables as dummy variables.
ordinal_features = ['LotShape','LandContour','Utilities','LandSlope','ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','HeatingQC','CentralAir','KitchenQual','Functional','GarageFinish','GarageQual','GarageCond']
le = LabelEncoder()
for feature in ordinal_features:
    train[feature]=le.fit_transform(train[feature])
#nominal features
non_nominal_features = ['Neighborhood','Exterior1st','Exterior2nd'] + ordinal_features

nominal_features = [feature for feature in cat_features if feature not in non_nominal_features]
nominal_features
```

```python
final_train = pd.get_dummies(data=train,columns=nominal_features) # encoded nominal variables as dummy variables and replaced in the training set
final_train.head(10)
```

# Training and evaluating

```python
X=final_train.drop(columns=['SalePrice'])
y=final_train['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=1)
reg = LinearRegression()
reg.fit(X_train,y_train)
coeff_df = pd.DataFrame(reg.coef_, X.columns, columns=['Coefficient'])
print("coefficients of the variables and the constant, of the linear equation","\n-----------------------------------------------------","\n-----------------------------------------------------")
print('Constant :',reg.intercept_)
coeff_df
 
```

```python
y_pred = reg.predict(X_test)
print('training score :', reg.score(X_train, y_train))
print('test score :', reg.score(X_test, y_test))
print('R^2 :',metrics.r2_score(y_test,y_pred))
print('Mean absolute Error',metrics.mean_absolute_error(y_test,y_pred))
```
